UV LF from the Donnan paper:

The evolution of the galaxy UV luminosity function at redshifts z  8 â€“ 15
from deep JWST and ground-based near-infrared imaging

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = np.linspace(-25, -16, 100)
```

```python
def phi(alpha, beta, phi_star, M_star):
    term_1 = 10**(0.4*(alpha+1)*(M - M_star))
    term_2 = 10**(0.4*(beta+1)*(M - M_star)) 
    return phi_star/(term_1+term_2)
     
```

z = 8

```python
alpha_1 = -2.04
beta_1 = -4.26
M_star_1 = -20.02
phi_star_1 = 3.30 *10**-4
```

z = 9

```python
alpha_2 = -2.10
beta_2 = -4.29
M_star_2 = -19.93
phi_star_2 = 2.10 *10**-4
```

z = 10.5

```python
alpha_3 = -2.10
beta_3 = -3.53
M_star_3 = -19.12
phi_star_3 = 3.32 *10**-4
```

z = 13.25

```python
alpha_4 = -2.10
beta_4 = -3.52
M_star_4 = -19.12
phi_star_4 = 0.51 *10**-4
```

```python
plt.plot(M, phi(alpha_1, beta_1, phi_star_1, M_star_1), '--', label = 'z = 8')
plt.plot(M, phi(alpha_2, beta_2, phi_star_2, M_star_2), label = 'z = 9')
plt.plot(M, phi(alpha_3, beta_3, phi_star_3, M_star_3), '--', label = 'z = 10.5')
plt.plot(M, phi(alpha_4, beta_4, phi_star_4, M_star_4), label = 'z = 13.5')
plt.yscale('log')
plt.ylim(1e-8, 1e-2)
plt.xlim(-23.5, -16)
plt.legend()
plt.xlabel('M$_{UV}$')
plt.ylabel('$\phi$ [mag$^{-1}$ Mpc$^{-3}$]')
```
