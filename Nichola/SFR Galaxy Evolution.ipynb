```python
# generating data for merger tree
```

```python
import numpy as np
import illustris_python as il
import matplotlib.pyplot as plt
```

```python
basePath = './sims.TNG/TNG100-1/postprocessing'
fields = ['SubfindID', 'SnapNum']
```

```python
tree = il.sublink.loadTree(basePath, 99, 3, fields=fields)
subfindIDs = tree['SubfindID']
snapnums = tree['SnapNum']
```

```python
subhalo_sfr = np.empty(len(snapnums))
subhalo_g_z = np.empty(len(snapnums))
subhalo_s_z = np.empty(len(snapnums))
```

```python
for snapshot in range(0, 100):
    path = './sims.TNG/TNG100-1/output/'
    fields = ['SubhaloGasMetallicitySfrWeighted', 'SubhaloStarMetallicityMaxRad', 'SubhaloSFR']
    subhalo = il.groupcat.loadSubhalos(path, snapshot, fields=fields)
    for index, snapnum in enumerate(snapnums):
        if snapnum == snapshot:
            subhalo_sfr[index] = subhalo['SubhaloSFR'][subfindIDs[index]]
            subhalo_g_z[index] = subhalo['SubhaloGasMetallicitySfrWeighted'][subfindIDs[index]]
            subhalo_s_z[index] = subhalo['SubhaloStarMetallicityMaxRad'][subfindIDs[index]]
    print(snapshot)
```

```python
redshifts = []
for snapshot in range(0, 100):
    path = './sims.TNG/TNG100-1/output/'
    redshifts.append(il.groupcat.loadHeader(path, snapshot)['Redshift'])
```

```python
import csv
file = open(f'./merger_rawdata', 'w')
writer = csv.writer(file)
writer.writerow(subhalo_sfr)
writer.writerow(subhalo_g_z)
writer.writerow(subhalo_s_z)
file.close()
```

```python
sfr = []
g_z = []
s_z = []
```

```python
for snapshot in range(0, 100):
    sum_sfr = []
    sum_g_z = []
    sum_s_z = []
    for index, snapnum in enumerate(snapnums):
        if snapnum == snapshot:
            sum_sfr.append(subhalo_sfr[index])
            sum_g_z.append(subhalo_g_z[index])
            sum_s_z.append(subhalo_s_z[index])
    sfr.append(np.sum(sum_sfr))
    g_z.append(np.sum(sum_g_z))
    s_z.append(np.sum(sum_s_z))
    print(snapshot)
```

```python
file = open(f'./merger_v3', 'w')
writer = csv.writer(file)
writer.writerow(sfr)
writer.writerow(g_z)
writer.writerow(s_z)
writer.writerow(redshifts)
file.close()
```
