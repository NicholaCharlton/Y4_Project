```python
import csv
import numpy as np
from hoki import load
import scipy.integrate
import illustris_python as il
import matplotlib.pyplot as plt
```

```python
basePath = './sims.TNG/TNG100-1/output/'
fields = ['SubhaloGasMetallicitySfrWeighted', 'SubhaloStarMetallicityMaxRad', 'SubhaloFlag', 'SubhaloSFR', 'SubhaloPos']
```

```python
Z_lower = [0.00000, 0.00005, 0.0005, 0.0015, 0.0025, 0.0035, 0.005, 0.007, 0.009, 0.012, 0.017, 0.025, 0.035]
Z_upper = [0.00005, 0.00050, 0.0015, 0.0025, 0.0035, 0.0050, 0.007, 0.009, 0.012, 0.017, 0.025, 0.035, 10.00]
```

```python
snapshots = np.linspace(1, 99, 99, endpoint=True)
```

```python
for snapshot_num in snapshots:
    subhalo = il.groupcat.loadSubhalos(basePath, snapshot_num, fields=fields)
    Z040_sfr_list = []
    Z030_sfr_list = []
    Z020_sfr_list = []
    Z014_sfr_list = []
    Z010_sfr_list = []
    Z008_sfr_list = []
    Z006_sfr_list = []
    Z004_sfr_list = []
    Z003_sfr_list = []
    Z002_sfr_list = []
    Z001_sfr_list = []
    Zem4_sfr_list = []
    Zem5_sfr_list = []
    cosmic_origin = []
    
    for subhaloID, Z in enumerate(subhalo['SubhaloStarMetallicityMaxRad']):
        if subhalo['SubhaloFlag'][subhaloID]:
            if 0 < subhalo['SubhaloPos'][subhaloID][0] < 7500 and 22577 < subhalo['SubhaloPos'][subhaloID][1] < 30077 and 14556.9 < subhalo['SubhaloPos'][subhaloID][2] < 22056.9:
                cosmic_origin.append([subhaloID, Z])

    for Z in cosmic_origin:
        if Z_lower[12] <= Z[1] < Z_upper[12]:
            Z040_sfr_list.append(float(subhalo['SubhaloSFR'][Z[0]]))
            cosmic_origin.remove([Z[0], Z[1]])
        
    for Z in cosmic_origin:
        if Z_lower[11] <= Z[1] < Z_upper[11]:
            Z030_sfr_list.append(float(subhalo['SubhaloSFR'][Z[0]]))
            cosmic_origin.remove([Z[0], Z[1]])
        
    for Z in cosmic_origin:
        if Z_lower[10] <= Z[1] < Z_upper[10]:
            Z020_sfr_list.append(float(subhalo['SubhaloSFR'][Z[0]]))
            cosmic_origin.remove([Z[0], Z[1]])
        
    for Z in cosmic_origin:
        if Z_lower[9] <= Z[1] < Z_upper[9]:
            Z014_sfr_list.append(float(subhalo['SubhaloSFR'][Z[0]]))
            cosmic_origin.remove([Z[0], Z[1]])
        
    for Z in cosmic_origin:
        if Z_lower[8] <= Z[1] < Z_upper[8]:
            Z010_sfr_list.append(float(subhalo['SubhaloSFR'][Z[0]]))
            cosmic_origin.remove([Z[0], Z[1]])
        
    for Z in cosmic_origin:
        if Z_lower[7] <= Z[1] < Z_upper[7]:
            Z008_sfr_list.append(float(subhalo['SubhaloSFR'][Z[0]]))
            cosmic_origin.remove([Z[0], Z[1]])
        
    for Z in cosmic_origin:
        if Z_lower[6] <= Z[1] < Z_upper[6]:
            Z006_sfr_list.append(float(subhalo['SubhaloSFR'][Z[0]]))
            cosmic_origin.remove([Z[0], Z[1]])
        
    for Z in cosmic_origin:
        if Z_lower[5] <= Z[1] < Z_upper[5]:
            Z004_sfr_list.append(float(subhalo['SubhaloSFR'][Z[0]]))
            cosmic_origin.remove([Z[0], Z[1]])
        
    for Z in cosmic_origin:
        if Z_lower[4] <= Z[1] < Z_upper[4]:
            Z003_sfr_list.append(float(subhalo['SubhaloSFR'][Z[0]]))
            cosmic_origin.remove([Z[0], Z[1]])
        
    for Z in cosmic_origin:
        if Z_lower[3] <= Z[1] < Z_upper[3]:
            Z002_sfr_list.append(float(subhalo['SubhaloSFR'][Z[0]]))
            cosmic_origin.remove([Z[0], Z[1]])
        
    for Z in cosmic_origin:
        if Z_lower[2] <= Z[1] < Z_upper[2]:
            Z001_sfr_list.append(float(subhalo['SubhaloSFR'][Z[0]]))
            cosmic_origin.remove([Z[0], Z[1]])
        
    for Z in cosmic_origin:
        if Z_lower[1] <= Z[1] < Z_upper[1]:
            Zem4_sfr_list.append(float(subhalo['SubhaloSFR'][Z[0]]))
            cosmic_origin.remove([Z[0], Z[1]])
        
    for Z in cosmic_origin:
        if Z_lower[0] <= Z[1] < Z_upper[0]:
            Zem5_sfr_list.append(float(subhalo['SubhaloSFR'][Z[0]]))
            cosmic_origin.remove([Z[0], Z[1]])
    
    s_Z040 = sum(Z040_sfr_list)/10**3
    s_Z030 = sum(Z030_sfr_list)/10**3
    s_Z020 = sum(Z020_sfr_list)/10**3
    s_Z014 = sum(Z014_sfr_list)/10**3
    s_Z010 = sum(Z010_sfr_list)/10**3
    s_Z008 = sum(Z008_sfr_list)/10**3
    s_Z006 = sum(Z006_sfr_list)/10**3
    s_Z004 = sum(Z004_sfr_list)/10**3
    s_Z003 = sum(Z003_sfr_list)/10**3
    s_Z002 = sum(Z002_sfr_list)/10**3
    s_Z001 = sum(Z001_sfr_list)/10**3
    s_Zem4 = sum(Zem4_sfr_list)/10**3
    s_Zem5 = sum(Zem5_sfr_list)/10**3
    
    total = [s_Z040, s_Z030, s_Z020, s_Z014, s_Z010, s_Z008, s_Z006, s_Z004, s_Z003, s_Z002, s_Z001, s_Zem4, s_Zem5]

    file = open(f'./limSFR_s{snapshot_num}_gas', 'w')
    writer = csv.writer(file)
    writer.writerow(total)
    file.close()
```
