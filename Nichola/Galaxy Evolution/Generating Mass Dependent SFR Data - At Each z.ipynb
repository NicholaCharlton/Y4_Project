```python
# finding subhalos in [snapshot] in particular mass bin

basePath_subs = './sims.TNG/TNG100-1/output/'
fields_subs = ['SubhaloFlag', 'SubhaloMass']

subhalo_list = []
subhalo = il.groupcat.loadSubhalos(basePath_subs, 4, fields=fields_subs)
for subhaloID, mass in enumerate(subhalo['SubhaloMass']):
    if subhalo['SubhaloFlag'][subhaloID]:
        if 0.1 < mass < 1:
            subhalo_list.append(subhaloID)
```

```python
# finding merger tree of above randomly chosen subhalo

basePath_tree = './sims.TNG/TNG100-1/postprocessing'
fields_tree = ['SubfindID', 'SnapNum']

tree = il.sublink.loadTree(basePath_tree, 4, sub, fields=fields_tree)
subfindIDs = tree['SubfindID']
snapnums = tree['SnapNum']

subhalo_sfr = np.empty(len(snapnums))
for snapshot in range(0, 100):
    fields = ['SubhaloSFRinMaxRad', 'SubhaloMass']
    subhalo = il.groupcat.loadSubhalos('./sims.TNG/TNG100-1/output/', snapshot, fields=fields)
    for index, snapnum in enumerate(snapnums):
        if snapnum == snapshot:
            subhalo_sfr[index] = subhalo['SubhaloSFRinMaxRad'][subfindIDs[index]]
    print(snapshot)
```
