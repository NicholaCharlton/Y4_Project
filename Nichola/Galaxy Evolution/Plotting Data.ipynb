```python
import csv
import h5py
from hoki.constants import *
import numpy as np
import scipy.integrate
import illustris_python as il
import matplotlib.pyplot as plt
import matplotlib.ticker as tck
from matplotlib.ticker import MaxNLocator
```

```python
data = []
with open('./merger_v2', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        data.append(row)
```

```python
sfr = []
g_z = []
s_z = []
redshifts = []
for index in range(0, 100):
    sfr.append(float(data[0][index]))
    g_z.append(float(data[1][index]))
    s_z.append(float(data[2][index]))
    redshifts.append(float(data[3][index]))
```

```python
plt.plot(redshifts, g_z, label='Gas SFR Weighted Metallicity')
plt.plot(redshifts, s_z, label='Stellar Metallicity')
#plt.yscale('log')
plt.ylabel('Metallicity')
plt.xlabel('Redshift')
plt.title('Metallicity Evolution of Merger Tree')
plt.legend(loc=1)
plt.show()
```

```python
plt.plot(redshifts, sfr)
plt.yscale('log')
plt.ylabel('Star Formation Rate Density ($M_\odot$ yr$^{-1}$ Mpc$^{-3}$)')
plt.xlabel('Redshift')
plt.title('SFR Evolution of Merger Tree')
plt.show()
```
